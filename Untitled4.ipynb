```python
#C:\Users\alexa\archive (1)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Airbnb NYC dataset
df = pd.read_csv('archive (1)/AB_NYC_2019.csv')
```

```python
# Preview the first few rows of the dataset
print(df.head())

# Check for missing values
print(df.isnull().sum())
```

```python
# Calculate the average reviews_per_month for each neighborhood_group
neighbourhood_group_avg_reviews = df.groupby('neighbourhood_group')['reviews_per_month'].transform('mean')

# Fill missing values in reviews_per_month with the average for the respective neighborhood
df['reviews_per_month'].fillna(neighbourhood_group_avg_reviews)

# Save the updated DataFrame back to a CSV file
df.to_csv('archive (1)/AB_NYC_2019.csv', index=False)


```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('archive (1)/AB_NYC_2019.csv')

# Ensure 'last_review' is treated as datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Calculate the average last_review date for each neighbourhood_group
average_last_review_per_neighbourhood = df.groupby('neighbourhood_group')['last_review'].mean()

# Display the average dates
print(average_last_review_per_neighbourhood)

# If you want to save the result to a new CSV file
average_last_review_per_neighbourhood.to_csv('archive (1)/AB_NYC_2019.csv', header=True)
```

```python
# Plot price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Price Distribution of NYC Airbnb Listings')
plt.xlabel('Price')
plt.show()

# Room types by neighborhood
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='neighbourhood_group', hue='room_type')
plt.title('Room Type Distribution by Neighborhood')
plt.show()
```

```python
from scipy import stats

# Example of hypothesis testing: Mean price difference between Manhattan and Brooklyn
manhattan_prices = df[df['neighbourhood_group'] == 'Manhattan']['price']
brooklyn_prices = df[df['neighbourhood_group'] == 'Brooklyn']['price']
t_stat, p_val = stats.ttest_ind(manhattan_prices, brooklyn_prices, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

```python
This code generates two scatter plots to visually analyze the geographic distribution of Airbnb listings in New York City, focusing on the neighbourhood group and the room type. Each plot provides insights into how listings are spread across the city in terms of these categories.

Plot 1: Neighbourhood Group Location
This scatter plot uses longitude and latitude to position each listing on a map-like layout.

Plot 2: Room Type Location per Neighbourhood Group
This scatter plot again maps each listing's geographic position but color-codes points based on the the room type (e.g. Entire home/apt, Private room).
This visualization highlights the types of accommodation available in different neighborhoods. Now it is easier to see where specific room types are concentrated.
```

```python
plt.ioff()

# Plot 1: Neighborhood Group Location
plt.figure(figsize=(10, 6))
sns.scatterplot(x='longitude', y='latitude', hue='neighbourhood_group', data=df)
plt.title('Neighbourhood Group Location')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
plt.close()  # Close the figure to free memory

#Room Type Location per Neighborhood Group
plt.figure(figsize=(10, 6))
sns.scatterplot(x='longitude', y='latitude', hue='room_type', data=df)
plt.title('Room Type Location per Neighbourhood Group')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
plt.close()  # Close the figure to free memory
```

```python
Geographical Distribution of Availability
This plot helps analyze availability. Now it is possible to potentially highlight areas with more or fewer available options throughout the year.
```

```python
plt.figure(figsize= (14,6))
plt.scatter(df.longitude, df.latitude, c = df.availability_365, alpha = 0.7,cmap ='summer',edgecolor = 'black')
c_bar = plt.colorbar()
c_bar.set_label('Availability')
plt.show()
```

```python
Comparison of Yearly Availability by Room Type¶
```

```python
plt.figure(figsize=(12,6))

result = df.groupby(["room_type"])['availability_365'].aggregate(np.median).reset_index().sort_values('availability_365')

sns.barplot(x='room_type', y="availability_365", data=df, order=result['room_type'])

plt.show()
```

```python

# Prepare data for the pie chart
labels = df['neighbourhood_group'].value_counts().index
sizes = df['neighbourhood_group'].value_counts().values

# Define shades of red
colors = ['#ff9999', '#ff4d4d', '#ff1a1a', '#e60000', '#b30000']  # Light to dark red
explode = [0] * len(labels)  # No explosion

# Create the pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Airbnb According to Neighbourhood Group', color='blue', fontsize=15)
plt.show()
```
